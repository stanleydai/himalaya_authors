```python
#looks like you can only get 600 results. fine. 
#for fun, what are the titles of these books?

for i in total_items:
    print(i['volumeInfo']['title'])
 
```

```python
#trying to return the authors for analysis 

authorlist = []

for i in total_items:
    try:
#         print(i['volumeInfo']['authors'])
        authorlist = authorlist + i['volumeInfo']['authors'] 
    except KeyError:
        pass

len(authorlist)
```

```python
authors = pd.DataFrame (authorlist, columns = ['name'])
#remove duplicates
authors = authors.drop_duplicates().reset_index(drop=True)
dfalldup = dfall.drop_duplicates(subset=['fname', 'lname'])
dfalldup.shape
#33326 unique people
pd.set_option('display.max_rows', None)

authors['lname'] = authors['name'].str.split().str[-1]

authors
```

```python
#match time! 

dfall['exactmatch'] = dfall.name.isin(authors.name).astype(int)

dfall.exactmatch.value_counts()

```

```python
#nicknames?
dfall['nickname'] = dfall.fname.str.extract('\(([^)]+)')
dfall['nickfullname'] = dfall['nickname'] + " " + dfall['lname']

#match time
dfall['nicknamematch'] = dfall.nickfullname.isin(authors.name).astype(int)

dfall.nicknamematch.value_counts()

```

```python
#try a last name match /filter. 

dfall['lmatch'] = dfall.lname.isin(authors.lname).astype(int)

dfall.lmatch.value_counts()
```

```python
#go through lname matches that aren't already a part of our finalmatch

dflast = dfall[(dfall['lmatch'] == 1) & (dfall['finalmatch']!=1)]

dflast.sort_values('lname')
```
